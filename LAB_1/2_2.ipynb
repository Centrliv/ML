```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
STEP = 0.1
x = np.arange(-5, 5, STEP, dtype='double')

f = x - (np.power(x, 3)/6) + (np.power(x, 5)/120) - (np.power(x, 7)/5040)
plt.plot(x, f)
plt.show()
```

```python
g = np.sin(x)
plt.plot(x, g)
plt.show()
```

```python
plt.plot(x, f, x, g)
```
