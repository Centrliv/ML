```python
import random
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

d = int(input())
total_points = 100

radius = d/2
samples = np.array([[random.uniform(-1*radius, radius), random.uniform(-1*radius, radius)] 
                         for i in range(total_points)])

theta = np.linspace(0, 2*np.pi, 100)
a = radius*np.cos(theta)
b = radius*np.sin(theta)

rect = patches.Rectangle((-1*radius, radius), d, -1*d, edgecolor='orange', facecolor='none' )

circle_points = []
square_points = []
for x,y in samples:
    is_point_in_circle = (x - 0)**2 + (y - 0)**2 <= radius**2
    if is_point_in_circle:
        circle_points.append(np.array([x, y]))
    else:
        square_points.append(np.array([x, y]))
circle_points_x = [p[0] for p in circle_points]
circle_points_y = [p[1] for p in circle_points]

square_points_x = [p[0] for p in square_points]
square_points_y = [p[1] for p in square_points]

figure, axes = plt.subplots(1)

axes.plot(a, b)
axes.add_patch(rect)
axes.scatter(circle_points_x, circle_points_y)
axes.scatter(square_points_x, square_points_y)
axes.set_aspect(1)

print('total_points', total_points)
print('points inside', len(circle_points))
print('fraction', len(circle_points)/100)
plt.show()
```
