```python
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import numpy as np
```

```python
df = pd.read_csv('housing.csv')
train_val, test = train_test_split(df, test_size=0.1)
train, val = train_test_split(train_val, test_size=0.1)
```

```python
train = pd.get_dummies(train, columns=['ocean_proximity']).reset_index(drop=True)
```

```python
imputer = SimpleImputer(strategy="median")
df_num = train.drop(columns=[c for c in train.columns if c.startswith('ocean_proximity')], axis=1)
imputer.fit(df_num)
X = imputer.transform(df_num)
tr = pd.DataFrame(X, columns=df_num.columns)
```

```python
train[train.total_bedrooms.isna()]
```

```python
tr = pd.concat([tr, train[[c for c in train.columns if c.startswith('ocean_proximity')]]], axis=1)
```

```python
tr['longitude'] = preprocessing.normalize([tr.longitude])[0]
tr['latitude'] = preprocessing.normalize([tr.latitude])[0]
```

```python
tr['average_rooms'] = tr.total_rooms / tr.households
tr['average_bedrooms'] = tr.total_bedrooms / tr.households
```
